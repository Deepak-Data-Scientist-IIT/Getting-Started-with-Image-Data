# **Loading Required Libraries and Images**

```
# Importing the required libraries
from skimage.io import imread, imshow
from skimage.filters import sobel_h,prewitt_h,prewitt_v,sobel_v
```

```
# Reading the image and plotting it
image = imread('dog.jpg', as_gray=True)
imshow(image, cmap= 'gray')
```

```
# Checking the shape of the image
image.shape
```

# **Sobel Kernel**

```
# Calculating the horizontal edges using sobel kernal
edges_sobel_horizental = sobel_h(image)
```

```
# Calculating the vertical edges using sobel kernal
edges_sobel_vertical = sobel_v(image)
```

```
# Ploting the sobel horizental edges
imshow(edges_sobel_horizental, cmap= 'gray')
```

```
# Plotting the sobel vertical edges
imshow(edges_sobel_vertical, cmap='gray')
```

# **Prewitt Kernal**

```
# Calculating horizontal edges using prewitt kernal
edges_prewitt_horizontal = prewitt_h(image)

# Calculating vertical edges using prewitt kernal
edges_prewitt_vertical = prewitt_v(image)
```

```
# Plotting the prewitt horizontal edges
imshow(edges_prewitt_horizontal, cmap= 'gray')
```

```
# Plotting the prewitt vertical edges
imshow(edges_prewitt_vertical, cmap= 'gray')
```
